```python
import pandas as pd
import numpy as np
from datascience import *
import random
from bs4 import BeautifulSoup

with open('data.txt', 'r') as myfile:
    data=myfile.read()

soup = BeautifulSoup(data, 'html.parser')
```

```python
blocks = soup.find_all('tr')
blocks
```

```python
def parse_block(b):
    name = b.find_all("td")[0].get_text()
    position = b.find_all("td")[1].get_text()
    department = b.find_all("td")[2].get_text()
    salary = b.find_all("td")[3].get_text()
    return {"name":name, "position":position, "salary":salary, "department":department, "salary":salary}
```

```python
def append_to_dataset(data, block):
    tab = pd.DataFrame.from_dict(parse_block(block), orient='index').T
    return data.append(tab)
data = pd.DataFrame(columns=['name', 'salary', 'position', 'department'])
for b in blocks[1:]:
    data = append_to_dataset(data, b)
```

```python
data = data.dropna()
data.salary = data.salary.apply(lambda x: x.replace('$', '').replace(',','')).astype(float)
```

```python
last_name = [name.split(", ")[0] for name in data.name]
first_name = [name.split(", ")[1] for name in data.name]
data['last_name'] = last_name
data['first_name'] = first_name
data.head()
```

```python
#first two letters of first name ansd first two of last name
anonymized = [first_name[i][:2]+last_name[i][:2] for i in range(len(first_name))]
data.name = anonymized
data.drop(['last_name', 'first_name'], axis=1, inplace=True)
data.head()
```

```python
data = data.drop_duplicates("name", keep='first')
data.head()
```

```python
data.to_csv("salaries.csv", index=False)
```
