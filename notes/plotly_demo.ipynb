```python
import plotly.plotly as py
import plotly.graph_objs as go
from datascience import *
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/adnanhemani/data8-assets/master/merged_election_data.csv").drop("Unnamed: 0", axis=1)
df.head()
```

```python
df_filtered = df[['votes_dem.16','votes_gop.16', "state_abbr"]]
new_df = df_filtered.groupby('state_abbr').sum()
new_df['code'] = new_df.axes[0]
new_df
```

```python
new_df['perc_diff'] = (new_df['votes_dem.16']-new_df['votes_gop.16'])/(new_df['votes_dem.16']+new_df['votes_gop.16'])
new_df.head()
```

```python
new_df['text'] = new_df['code']+'<br>'+ "Democrats " + new_df['votes_dem.16'].astype('str')+'<br>'+ "Republicans " + new_df['votes_gop.16'].astype('str')
```

```python
new_df[new_df['code']=='AK']
```

```python
for col in new_df.columns:
    new_df[col] = new_df[col].astype(str)

scl = [[-0.5, 'rgb(242,240,247)'],[-0.3, 'rgb(218,218,235)'],[-0.1, 'rgb(188,189,220)'],\
            [0.1, 'rgb(158,154,200)'],[0.3, 'rgb(117,107,177)'],[0.5, 'rgb(84,39,143)']]


data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = new_df['code'],
        z = new_df['perc_diff'].astype(float),
        locationmode = 'USA-states',
        text = new_df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Percentage Difference")
        ) ]

layout = dict(
        title = 'My Title',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```
